cube transform

```python
#Copying the DataFrame
transforms = data.copy()
#Adding a new column cube to the dataframe
transforms['cube'] = transforms['calories'].transform(lambda x: np.power(x, 3))
#plots a histogram of the values in the cube column
transforms['cube'].hist()
```

iV) Normalize the "age", "height", and "weight", and Standardize "steps" and "heart rate" columns in a separate column at the end of the dataframe

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
#Copying the DataFrame
data_copy = data.copy()
#Scaling the Features
scaler = MinMaxScaler()
standard = StandardScaler()
#scaling to the Age, Height, and Weight columns
data_copy[['Age', 'Height', 'Weight']] = scaler.fit_transform(data_copy[['age', 'height', 'weight']])
#scaling to the heartrate and steps columns
data_copy[['heartrate', 'Steps']] = standard.fit_transform(data_copy[['hear_rate', 'steps']])
#prints the transformed data
print(data_copy[['Age', 'Height', 'Weight']].head())
print(data_copy[['heartrate', 'Steps']].head())
#saving the updated dataframe
data_copy.to_csv("modifiedawfbdata.csv", index=False)
print("The updated DataFrame has been saved to 'modifiedawfbdata.csv'.")
```
